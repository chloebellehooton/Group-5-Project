```python
#Created COLUMN MEDIAN OF TMAX FOR EACH MONTH
MED= wthr_Siskiyou_df.groupby(['DATE'])[['TMAX']].apply(np.nanmean)
MED.name = 'MEDIAN_TMAX'
wthr_Siskiyou_df=wthr_Siskiyou_df.join(MED, on=['DATE'])
wthr_Siskiyou_df.head()
```

```python
avgprcp = wthr_Siskiyou_df.groupby(['DATE'])[['PRCP']].apply(np.nanmean)
avgprcp.name = 'MEAN_PRCP'
wthr_Siskiyou_df=wthr_Siskiyou_df.join(avgprcp, on=['DATE'])
wthr_Siskiyou_df.head()
```

```python
#Group the DF by County & Date, then calculate 
SUM = wthr_Siskiyou_df.groupby(['DATE'])[['PRCP']].apply(np.nansum)
SUM.name = 'SUM_PRCP'
wthr_Siskiyou_df = wthr_Siskiyou_df.join(SUM, on=['DATE'])
wthr_Siskiyou_df.head()
```

# Loading the wildfire data


```python
file_path = "../Resources/wildfire_clean.csv"
wildfire_df=pd.read_csv(file_path,index_col=0)
print(wildfire_df.shape)
wildfire_df.head()
```

```python
wildfire_df.info()
```

```python
# Removing the columns which has same data or large number of null/NaN values
n_cols = [i for i in wildfire_df.columns if i not in ('CountyIds', 'CrewsInvolved','Fatalities','Name','PersonnelInvolved','StructuresDamaged','StructuresDestroyed')]
new_wildfire_df = wildfire_df[n_cols]
print(new_wildfire_df.shape)
new_wildfire_df
```

```python
new_wildfire_df.Counties.unique()
```

```python
#Removing space from county names to make it par with weather dataframe
new_wildfire_df["Counties"] = new_wildfire_df["Counties"].str.replace(" ","")
new_wildfire_df.head()
```

```python
new_df = pd.merge(new_wildfire_df, mod_wthr_df,  how='left', left_on=['Counties','Year-Month'], right_on = ['COUNTY','Year-Month'])
print(new_df.shape)
new_df.head(10)
```

```python
# Null/NaN Value count in columns
for column in new_df.columns:
    print(f"Column {column} has {new_df[column].isnull().sum()} null values")
```

```python
nan_df=new_df[new_df.isna().any(axis=1)]
```

```python
nan_df.Counties.unique()
```

```python
# Remove rows that have at least 1 null value.
# YOUR CODE HERE
wildfire_wthr_df=new_df.dropna()
print(wildfire_wthr_df.shape)
wildfire_wthr_df.head(10)
```

```python
#Creating new merged df with only revelant columns
f_cols = [i for i in wildfire_wthr_df.columns if i not in ('ArchiveYear','Latitude','Longitude','Started','MAXIM_TMAX','SUM_PRCP','Year_x','Year_y','Month_y')]
wildfire_wthr_df = wildfire_wthr_df[f_cols]
wildfire_wthr_df.head()
```
