```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
file_path = "Resources/wildfire_wthr.csv"
wildfire_wthr_df=pd.read_csv(file_path)
print(wildfire_wthr_df.shape)
wildfire_wthr_df.head()
```

```python
wildfire_wthr_df=wildfire_wthr_df.drop(columns=['MEAN_PRCP'])
print(wildfire_wthr_df.shape)
wildfire_wthr_df.head()
```

```python
#Setting Unique id and year month as index
wildfire_wthr_df.set_index(["UniqueId","Year-Month","Counties"],inplace=True, drop=True)
print(wildfire_wthr_df.shape)
wildfire_wthr_df.head(20)
```

```python
# Use get_dummies() to create variables for text features.
# YOUR CODE HERE
# Binary encoding using Pandas (multiple columns)
X = pd.get_dummies(wildfire_wthr_df, columns=["COUNTY","Month_x"])
X.head()
```

```python
X_scaled=StandardScaler().fit_transform(X)
print(X_scaled[0:5])
```

```python
# Initialize PCA model
pca = PCA(n_components=3)
```

```python
X_pca = pca.fit_transform(X_scaled)
print(X_pca)
```

```python
#pcs_df = pd.DataFrame(index=X.index,data=X_pca,columns=["PC 1","PC 2","PC 3","PC 4","PC 5"])
pcs_df = pd.DataFrame(index=X.index,data=X_pca,columns=["PC 1","PC 2","PC 3"])
pcs_df.head(10)
```

```python
inertia = []
k = list(range(1, 11))
```

```python
# Looking for the best K
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
```

```python
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

```python
# Initialize the K-Means model.
# YOUR CODE HERE
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
# YOUR CODE HERE
model.fit(pcs_df)


# Predict clusters
predictions = model.predict(pcs_df)

# YOUR CODE HERE
predictions
```

```python
# Create a new DataFrame including predicted clusters and wildfire_wthr features.
# Concatentate the wildfire_wthr and pcs_df DataFrames on the same columns.
# YOUR CODE HERE

results= pd.concat([X, pcs_df], axis=1, join='inner')
print(results.shape)
results.head()
```

```python
#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
# YOUR CODE HERE
results["Class"] = model.labels_
print(results.shape)

results.head(10)
```

```python
v_cols=["AcresBurned","MEDIAN_TMAX","cum_prcp"]  
X_MinMax =  results[v_cols]   
X_MinMax
```

```python
# Scaling data to create the scatter plot 
 # Scale the dataset using MinMaxScaler()
X_Min_scaled = MinMaxScaler().fit_transform(X_MinMax)
X_Min_scaled
```

```python
# Create a new DataFrame that has the scaled data with the results DataFrame index.
# YOUR CODE HERE
new_results  = pd.DataFrame(index=results.index,data=X_Min_scaled,columns=["AcresBurned","MEDIAN_TMAX","cum_prcp"])
new_results.head(10)
```

```python
new_results =  pd.concat([new_results, results["Class"]], axis=1, join='inner')
print(new_results.shape)
new_results.head(10)
```
