```python
#Import dependencies
import pandas as pd
import glob
import os
```

### Import & Cleanse Wildfire Data

```python
#  Import and read the Wildfire CSV.
wildfire_df = pd.read_csv("Resources/California_Fire_Incidents.csv")
wildfire_df.head()
```

```python
#Drop excess columns
wildfire_df = wildfire_df.drop(['Active','AdminUnit','AirTankers','CalFireIncident','CanonicalUrl',
                                'ConditionStatement','ControlStatement','Dozers','Engines','Extinguished',
                                'Featured','Final','FuelType','Helicopters','Injuries','Location',
                                'MajorIncident','PercentContained','Public','SearchDescription','SearchKeywords',
                                'Status','StructuresEvacuated','StructuresThreatened','Updated','WaterTenders'],1)
wildfire_df.head()
```

```python
#Confirm datatypes for each column
wildfire_df.info()
```

```python
#Drop duplicates based on UniqueID
wildfire_df.drop_duplicates("UniqueId", inplace = True)
wildfire_df.head()
```

```python
#Check for null data
wildfire_df.count()
```

```python
#Drop null data from AcresBurned
wildfire_df = wildfire_df.dropna(subset = ['AcresBurned'])
wildfire_df.head()
```

```python
#Check for null data
wildfire_df.count()
```

### Import & Merge County Data

```python
#Import multiple CSV files into 1 dataframe
path= os.path.join("Resources", "CountyWeather")
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df['source'] = os.path.basename(filename)
    li.append(df)

county_df = pd.concat(li, axis=0, ignore_index=True)
county_df.head()
```

```python
#Drop excess columns
county_df = county_df.drop(['ELEVATION','PRCP_ATTRIBUTES','TAVG_ATTRIBUTES','TMAX_ATTRIBUTES',
                            'TMIN_ATTRIBUTES','AWND_ATTRIBUTES'],1)
county_df.tail()
```

```python
#Confirm datatypes for each column
county_df.info()
```

```python
#Check for null data
county_df.count()
```

### Calculate monthly precipitation

```python
#Calculate monthly precipitation (sum all stations per county)
county_df_prcp = county_df.groupby(['source','DATE']).sum()
county_df_prcp
```

```python
#Drop all columns except source, date, prcp
county_df_prcp = county_df_prcp.drop(['LATITUDE','LONGITUDE','TAVG','TMAX',
                            'TMIN','AWND'],1)
county_df_prcp = county_df_prcp.rename(columns={'PRCP': 'MONTHLY_PRCP'})
county_df_prcp.tail()
```

```python
#Add monthly precipitation to county_df
test_df = pd.merge(county_df, county_df_prcp,how='left', 
                   left_on=['source','source'], right_on = ['DATE','DATE'])
test_df.head()
```
